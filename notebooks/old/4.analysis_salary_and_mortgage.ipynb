## 1. Mortgage Cost Calculation

To continue our study and calculate how long it will take the residents of Barcelona to pay off a mortgage, or what percentage of their salary they will need to allocate to paying a mortgage over 30 years, we need to define how we calculate the mortgage amount, as it is not straightforward.

We will need a function that, **given the price of the property, the salary, and the interest rate, can tell us how long the person will need to pay it off**, given that only **35%** of the salary should be spent on the mortgage, and substracting the **20% of the total amount**, since we **assume is the initial payment, and it is already paid**.

If we were to simply divide the total amount by the portion of the salary that would need to be allocated to the mortgage each month, we would have a very poor estimate of how long it would take, as **the total amount of the mortgage will directly depend on the time it takes to pay it off**. Therefore, we see that the calculation is not that simple, which is why we go through the entire process outlined below, including the following **formula**:

```python
from IPython.display import display, Math # type: ignore

display(Math(r'Total Mortgage Amount = AmountMortgage + TotalInterests'))
display(Math(r'AmountMortgage = FlatValue + Taxes - InitialPayment = FlatValue + FlatValue*0.105 - FlatValue*0.2'))
display(Math(r'Total Interests = \sum_{i=0}^{nyears}((A_m - \frac{A_m}{nyears}*i)*interest_{rate}) = \frac{nyears}{2} * (FirstInterest + LastInterest) = \frac{nyears}{2} * ((A_m + A_y)*interest_{rate})'))
```

Here is a simple way to calculate the amount of a mortgage. Basically, **each year** you will pay **the corresponding portion of the total value divided by the number of years you've agreed to**, **plus the interest percentage on the remaining amount you still owe**.

```python
# calculation of the total amount of a Mortgage through the years

# inputs
n_years = 25
tax = 0.03
initial_imp = 300000
imp = initial_imp + (initial_imp*0.105) - (initial_imp*0.2)

# imp paid per year
imp_year = imp / (n_years)
taxes = []
print("Imp per year: ", imp_year)

# calculating the extra import per year given the tax
for i, year in enumerate(range(0,n_years)):
    print(f"Year: {year}, Taxes: {(imp-imp_year*year)*tax:.1f}, Remaining: {imp-imp_year*year:.1f}")
    taxes.append((imp-imp_year*year)*tax)

# printing the total amount the mortgage will cost
total_payment = np.sum(taxes) + imp
print("Total payment in taxes =", np.sum(taxes), "| Total payment mortgage =", total_payment)
```

Here we have a simplified calculation of the above algorithm. This corresponds to the formula mentioned above.
Also we add the calculation if a person can afford the Mortgage with his salary.

```python
n_years = 30
tax = 0.03
initial_imp = 300000
imp = initial_imp + (initial_imp*0.105) - (initial_imp*0.2)

net_salary = 50000
mortgage_salary = net_salary * 0.35

imp_year = imp / n_years
total_taxes = (n_years/2)*((imp*tax)+(imp_year*tax))
total_payment = imp + total_taxes

print("Total payment in taxes =", total_taxes, "| Total payment mortgage =", total_payment)
print("Total payment per year =", total_payment / n_years, "| Affordable salary =", mortgage_salary)
print("Can pay?", mortgage_salary > total_payment / n_years)
```

### 1.1. Calculation of the optimal years to pay the mortgage depending on the salary

Here is where we create the algorithm that we are going to use. Basically, the algorithm calculates the **final amount of the property depending on the duration of the mortgage**, the final amount, and the interest rate. The algorithm iterates through the number of years until **it finds a minimum duration where the monthly payment is less than 35% of the monthly salary**, as this is the recommended threshold.

```python
# inputs
tax = 0.03
initial_imp = 475000
imp = initial_imp + (initial_imp*0.105) - (initial_imp*0.2)
print("Imp before taxes =",imp)

net_salary = 50000
mortgage_salary = net_salary * 0.35

# iterating through the number of years to pay the mortgage
n_years_list = np.arange(2,71)
payment_per_year_list = []
total_taxes_year_list = []
minimum_year_affordable = 70

for i, n_years in enumerate(n_years_list, start=2):
    imp_year = imp / n_years
    total_taxes = (n_years/2)*((imp*tax)+(imp_year*tax))
    total_payment = imp + total_taxes
    payment_per_year = total_payment / n_years
    payment_per_year_list.append(payment_per_year)
    total_taxes_year_list.append(total_taxes)
    
    # check if the total amount is affordable by the salary.
    if (payment_per_year <= mortgage_salary) & (i<minimum_year_affordable):
        minimum_year_affordable = i
        payment_affordable = payment_per_year
        minimum_taxes_pay = total_taxes

print("Minimum taxes to pay =", minimum_taxes_pay)
print("Total payment =", imp+minimum_taxes_pay)
print("Anual salary to allocate towards the mortgage =", mortgage_salary)
print("Minimum years to pay =", minimum_year_affordable)
print("Affordable payment per year =", payment_affordable)

plt.plot(n_years_list, payment_per_year_list, 'bo', label='payment per year')
plt.plot(minimum_year_affordable, payment_affordable, 'ro', label='optimal years can afford')
plt.plot(n_years_list, total_taxes_year_list, 'yo', label='taxes')
plt.legend()
plt.title("price per year depending on number of mortgage years")
plt.xlabel("Mortgage years")
plt.ylabel("Price per year")
plt.show()
```

As we can see, the longer we take to pay off the mortgage, the less we will pay per year, but the interest will increase significantly.

```python
# defining the function that will calculate the above algorithm
def get_years_mortgage(initial_imp, net_salary, interest_rate):
    '''
    inputs:
     - initial_imp: Amount of the apartment as advertised
     - Annual net salary
     - Interest rate
    outputs:
     - minimum years to pay the mortgage
     - total value of the flat not considering the initial payment
    '''
    imp = initial_imp + (initial_imp*0.105) - (initial_imp*0.2)
    mortgage_salary = net_salary * 0.35
    n_years_list = np.arange(2,120)
    minimum_year_affordable = 120

    for i, n_years in enumerate(n_years_list, start=2):
        imp_year = imp / n_years
        total_taxes = (n_years/2)*((imp*interest_rate)+(imp_year*interest_rate))
        total_payment = imp + total_taxes
        payment_per_year = total_payment / n_years
        
        if (payment_per_year <= mortgage_salary) & (i<minimum_year_affordable):
            minimum_year_affordable = i
            minimum_taxes_pay = total_taxes
            # break;
            return minimum_year_affordable, imp+minimum_taxes_pay
            
    # return minimum_year_affordable, imp+minimum_taxes_pay

get_years_mortgage(475000, 50000, 0.03)
```

## 2. Years to pay mortgage results

We calculate how many years will be needed to pay off each property, **assuming that the person buying the property lives in and earns the average salary of the census section where the property is located.** Two amounts are calculated, depending if the mortgage is done by only one person or a couple.

The interest rate we use is the Euribor, as it is the average interest rate of European banks and the reference rate for most mortgages.

```python
euribor_pct = 0.03

gdf_flats_with_salaries[["years_mortgage","total_mortgage_amount"]] = gdf_flats_with_salaries.apply(lambda row: pd.Series(get_years_mortgage(row["value"], row["salary"], euribor_pct)), axis=1)
gdf_flats_with_salaries[["years_mortgage_couple","total_mortgage_amount_couple"]] = gdf_flats_with_salaries.apply(lambda row: pd.Series(get_years_mortgage(row["value"], row["salary"]*2, euribor_pct)), axis=1)
```

```python
print(f"Barcelona Average Mortgage Years (one person) {np.mean(gdf_flats_with_salaries.years_mortgage):.1f}")
print(f"Barcelona Average Mortgage Years (couple) {np.mean(gdf_flats_with_salaries.years_mortgage_couple):.1f}")
print(len(gdf_flats_with_salaries))

avg_years_census_section = gdf_flats_with_salaries[["census_section_code","years_mortgage","years_mortgage_couple", "value","salary"]].groupby("census_section_code").mean().sort_values("years_mortgage", ascending=False)
avg_years_level8 = gdf_flats_with_salaries[["level8","years_mortgage","years_mortgage_couple", "value","salary"]].groupby("level8").mean().sort_values("years_mortgage", ascending=False).reset_index(drop=False)
avg_years_level7 = gdf_flats_with_salaries[["level7","years_mortgage","years_mortgage_couple", "value","salary"]].groupby("level7").mean().sort_values("years_mortgage", ascending=False)
avg_years_level7
```

The results speak for themselves. On average, a person will need to dedicate 65 years to paying off a mortgage if they want to buy a property on their own. If they do so as a couple, the average drops to 45 years, assuming steady salary growth from 2021 and that both people in the couple earn the same amount.

However, it's important to note one crucial thing: we are assuming that people's salaries will remain the same over the years, which is not the case. Therefore, that 35% of the salary will increase over time, likely allowing the mortgage to be paid off in less time.

We are also assuming a fixed-rate mortgage at a time when interest rates are very high. In the future, interest rates are expected to be lower.

However, even though this might offer some hope, banks still base mortgage approvals on your current salary.

This also explains why the vast majority of properties today are purchased outright, without a mortgage, either by wealthy individuals or investors.

In conclusion, it is very challenging for people in Barcelona to buy a property in their city unless they have significant additional income beyond their regular job. If there isn't some change, people will have to allocate a higher percentage of their salary (which we'll explore in the next section).

```python
from fuzzywuzzy import process

def merge_geodf_with_data(geo_df, column_geodf, data, column_data):
    unique_values_df1 = geo_df[column_geodf].unique()
    unique_values_df2 = data[column_data].unique()

    matches = {}
    for value in unique_values_df1:
        matches[value] = process.extractOne(value, unique_values_df2)[0] #find_most_similar(value, unique_values_df2)
    
    matches_df = pd.DataFrame(matches.items(), columns=[column_geodf, column_data])
    
    my_geo_df_names = pd.merge(geo_df, matches_df, on=column_geodf)
    merged = pd.merge(my_geo_df_names, data, on=column_data)

    return merged

# merging the GeoDataFrame with our data by Neighbourhood name
gdf_avg_years_level8 = merge_geodf_with_data(my_geo_df, "nom_barri", avg_years_level8, "level8")
```

```python
import folium
from IPython.display import display
import branca.colormap as cm

# minimum and maximum values to set the colormap range
min_value = gdf_avg_years_level8['years_mortgage'].min()
max_value = gdf_avg_years_level8['years_mortgage'].max()

# create our map and colormap to plot
m = folium.Map([41.3851, 2.1734], zoom_start=12, tiles="cartodbpositron")

# going through all neighbourhoods and adding the geometrical and statistical information to the map
colormap = cm.LinearColormap(['green', 'red'], vmin=min_value, vmax=max_value)
for index, row in gdf_avg_years_level8.iterrows():
    color = colormap(row["years_mortgage"])
    polygon = folium.GeoJson(row["geometry"], style_function=lambda x, color=color: {'fillColor': color, 'color': color, 'weight': 1, 'fillOpacity': 0.7})
    polygon.add_child(folium.Popup(f'{row["nom_barri"]} - {row["years_mortgage"]:.1f} years'))
    m.add_child(polygon)

    folium.Marker(
        location=[row.geometry.centroid.y, row.geometry.centroid.x],
        icon=folium.DivIcon(html=f'<div style="font-size: 6pt; color: black;">{row["years_mortgage"]:.1f}</div>')
    ).add_to(m)

colormap.caption = 'Mortgage Years'
m.add_child(colormap)

display(m)
```

## Amount to fix a 30-years mortgage

As we mentioned earlier, in this final section we will look at what percentage of their salary people living in Barcelona would need to dedicate to buy an apartment in their neighborhood with a 30-year mortgage.

As we know, **we used the salary by census section and the value of each flat to make the calculations, and the neighborhood corresponding to each flat, along with the geometry of the neighborhood, to plot the results**.

In this case, calculating the cost of the mortgage is not too complicated, since we have a fixed value for the years.

```python
def get_mortgage_payment_year(initial_imp, n_years, interest_rate):
    imp = initial_imp + (initial_imp*0.105) - (initial_imp*0.2)
    total_taxes = (n_years/2)*((imp*interest_rate)+(imp_year*interest_rate))
    total_payment = imp + total_taxes
    payment_per_year = total_payment / n_years
    return payment_per_year
```

```python
gdf_flats_with_salaries["mortgage_payment_per_year_30"] = gdf_flats_with_salaries.apply(lambda row: pd.Series(get_mortgage_payment_year(row["value"], 30, 0.03)), axis=1)
gdf_flats_with_salaries["pct_paytment_mortgage_over_anual_salary"] = gdf_flats_with_salaries["mortgage_payment_per_year_30"]/gdf_flats_with_salaries["salary"]
gdf_flats_with_salaries["pct_paytment_mortgage_over_anual_salary_couple"] = gdf_flats_with_salaries["mortgage_payment_per_year_30"]/(gdf_flats_with_salaries["salary"]*2)
```

```python
yearly_payment_per_hood = gdf_flats_with_salaries[["level8","pct_paytment_mortgage_over_anual_salary", "pct_paytment_mortgage_over_anual_salary_couple"]].groupby("level8").mean().reset_index()
print(f"Barcelona Average Salary Pergentage to afford a 30-year Mortgage: {np.mean(gdf_flats_with_salaries.pct_paytment_mortgage_over_anual_salary)*100:.1f}% (one person)")
print(f"Barcelona Average Salary Pergentage to afford a 30-year Mortgage: {np.mean(gdf_flats_with_salaries.pct_paytment_mortgage_over_anual_salary_couple)*100.:.1f}% (couple)")
avg_percentage_salary_level8 = gdf_flats_with_salaries[["level8","pct_paytment_mortgage_over_anual_salary", "pct_paytment_mortgage_over_anual_salary_couple"]].groupby("level8").mean().sort_values("pct_paytment_mortgage_over_anual_salary", ascending=False)
avg_percentage_salary_level7 = gdf_flats_with_salaries[["level7","pct_paytment_mortgage_over_anual_salary", "pct_paytment_mortgage_over_anual_salary_couple"]].groupby("level7").mean().sort_values("pct_paytment_mortgage_over_anual_salary", ascending=False)
display(avg_percentage_salary_level7)
```
